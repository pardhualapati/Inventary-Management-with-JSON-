```python
import json
import time

#reading the inventery
fd = open('Records.json','r')
js = fd.read()
fd.close()


#converting the json to dict
record = json.loads(js)

print("--------------- Menu -----------------\n")

#Displaying the inventery to user
for key in record.keys():
    print(key," : ", record[key]['Name']," | ",record[key]['Price'],"Rs"," | ",record[key]['Quantity'])


#taking details 
ui_name = str(input("Enter your Name: "))
ui_mail = str(input("Enter mailid: "))
ui_phno = str(input("Enter Phno: "))
ui_id   = str (input("Enter Product ID: "))
ui_qn   = int(input("Enter Quantity: "))


#if we have suffient quanity -- our inventery quantity is greater than the user required quantity -Generating bill
if(ui_qn < record[ui_id]['Quantity']):
    
    print("----------------------------------------")
    print("                 Bill                 \n")

    print("Product Name : ",record[ui_id]['Name'])
    print("Price (Rs)   : ",record[ui_id]['Price'])


    print("---------------------------------------\n")
    print("Billing Cost : ",record[ui_id]['Price']* ui_qn,"Rs")
    print("-----------------------------------------")
    
    #concatinating the details of user
    sales = ui_name+','+ui_mail+','+ui_phno+','+ui_id+','+str(ui_qn)+','+str(record[ui_id]['Price']* ui_qn)+','+time.ctime()+'\n'
    
    #Generating sales file 
    fd = open('sales.txt','a')
    fd.write(sales)
    fd.close()
    
    #After Generating the sales file updating the inventery!
    record[ui_id]['Quantity'] = record[ui_id]['Quantity'] - ui_qn

    
# we have zero quantity in the inventery we can't sale anything.
elif(record[ui_id]['Quantity'] == 0):
    print('We dont have the stock right now')
    
    
#we have less quantity than the user required quantity    
else:
    
    print("Sorry we only have",record[ui_id]['Quantity'],record[ui_id]['Name'])
    print("Would you like to purchase?")
    print("-----------------------------------------")
    ch = input("Press Y/y to purchase : ")
    print("-----------------------------------------")
    
    if(ch == 'Y' or ch == 'y'):
        
        print("----------------------------------------")
        print("                 Bill                 \n")

        


        print("Product Name : ",record[ui_id]['Name'])
        print("Price (Rs)   : ",record[ui_id]['Price'])


        print("---------------------------------------\n")
        print("Billing Cost : ",record[ui_id]['Price']* record[ui_id]['Quantity'],"Rs")
        print("-----------------------------------------")
        
        #concatinating the details of user
        sales = ui_name+','+ui_mail+','+ui_phno+','+ui_id+','+str(record[ui_id]['Quantity'])+','+str(record[ui_id]['Price']* record[ui_id]['Quantity'])+','+time.ctime()+'\n'
        
        #Generating sales file 
        fd = open('sales.txt','a')
        fd.write(sales)
        fd.close()
    
        #updating the inventery
        record[ui_id]['Quantity'] = 0
        
    else:
        print('Thanks for your time - We will get the stock soon!!')


#converting the dict to json format to update the inventery
js = json.dumps(record)


#updating the inventery.
fd = open('Records.json','w')
fd.write(js)
fd.close()
print("Inventery Updated")
```
